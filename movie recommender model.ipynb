```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):#taking first 3 cast members only
    counter=0
    L=[]
    for i in ast.literal_eval(obj):
        if counter!=3:
         L.append(i['name'])
         counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetchdirector(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
              L.append(i['name'])
              break  
    return L
```

```python
movies['crew']=movies['crew'].apply(fetchdirector)
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    string=' '.join(y)
    return string
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title'] == movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list :
        print (i[0])
```

```python
recommend('Avatar')
```
